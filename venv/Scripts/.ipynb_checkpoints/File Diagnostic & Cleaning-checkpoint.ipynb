# Data Diagnostic
This package is used to analyze the compatibility of files generated from ALD machine. 
The methods include:

<b>Determining empty files vs valid files</b>
* Some files were generated but has no information aside from header data

<b>Identifying the characteristics of each file</b>
* Some files have too many columns than actually used, so we have to rearrange so that the header match the appropriate data.

<b>Determining set of files from different folders that share common names</b>
* 5 Files should be generated when ALD machine is run, but this is not always the case.

# A. File size diagnostics

```python
# Get file size distribution
def plot_loghist(x, bins):
      hist, bins = np.histogram(x, bins=bins)
      logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
      plt.hist(x, bins=logbins)
      plt.xscale('log')

def getfileSizeDistribution (*folder_name):
    for name in folder_name:
        base_path =r'C:\Users\Tanat\PycharmProjects\nhi-datacollection-ald\MachineSide\_Logfile'
        path = r''.join([base_path, '\\', name])
        print("Folder: " + name)
        filenames = glob.glob(path + "/*.txt")

        size_array = []

        for filename in filenames:
            if os.path.getsize(filename) > 0:
                size_array.append(os.path.getsize(filename)/1000)

        print("Mean: " + str(round(statistics.mean(size_array),2)) + " MB")
        print("Median: " + str(round(statistics.median(size_array),2)) + " MB")
        print("Mode: " + str(round(statistics.mode(size_array),2)) + " MB")
        
        plot_loghist(size_array, 10)
        
        plt.title('File size distribution for '+str(name))
        plt.xlabel('File size (Log scale)')
        plt.ylabel('count')
        # plt.figure(figsize=(20, 20), dpi= 320)
        #plt.savefig('File size distribution'+str(name)+'.png')
        #plt.clf()
        print("*"*15) 
    return

getfileSizeDistribution("Heater Data")
```

# B. Filer Empty / Valid Files 
In addition to determining whether the file is empty, this section indicates the number of columns in the data (first line) and index the integer number on the row. This is important as the only number in Heater Data that was recorded as integer is the number of cycles remaining. We can use that to re-issue the header.

```python
def filter_valid(element):
    invalid_entry = ['nan', None, 'NA']

    if(str(element) in invalid_entry):
        return False  # Discount the invalid
    else:
        return True  # Count the valid

def list_find_integer(given_list):  
    # The only number in Heater Data that was recorded as integer 
    # is number of cycles remaining, and always available.
    ndx = 0
    last_ndx = 0
    for word in given_list:
        if str(word).isdigit():
            break #last_ndx = ndx
        ndx += 1
    return ndx  # Return the first integer

def count_empty_dataframe(folder_name):
    # Note the files that doesn't have log data (only headers are available)
    # get data file names
    base_path =r'C:\Users\Tanat\PycharmProjects\nhi-datacollection-ald\MachineSide\_Logfile'
    path = r''.join([base_path, '\\', folder_name])
    print("path: " + path)
    filenames = glob.glob(path + "/*.txt")

    counter = 0
    anticounter = 0
    column_count = list()
    int_count = list()
    iterable = list()
    coll_count = ''
    int_counter = ''
    
    output_filename1 = "".join(["empty_files_", folder_name.replace(" ", ""), ".txt"])
    output_filename2 = "".join(["valid_files_", folder_name.replace(" ", ""), ".txt"])
    
    with open(output_filename1, "w+") as file1:
        with open(output_filename2, "w+") as file2:
            for filename in filenames:
                try:# In case that file is EMPTY
                    
                    '''if not '2019_06_25-13-46_Optimized 200C.txt' in filename:
                        raise ValueError()  # target specific file
                    else:
                        print(filename)'''
                    
                    
                    data = pd.read_csv(filename, sep="\t", engine='python', nrows=1)
                    if data.empty:
                        counter += 1
                        file1.write(filename[:]+"\n")
                    else:  # In case that file is NOT EMPTY
                        anticounter += 1
                        iterable = data.values.tolist()
                        cleanedList = list(filter(filter_valid, iterable[0]))
                                        # Another quick way to create a filter is to use an in-line conditional expression below:
                                        # cleanedList = [x for x in iterable[0] if str(x) != 'nan']
                        k = len(cleanedList)  # get the number of elements in the cleanedlist
                        int_index = list_find_integer(cleanedList)
                        file2.write(filename[74:]+"\t"+str(k)+"\t"+str(int_index)+"\n")  # Record the valid filenames + number of column  

                        if int_index not in int_count:
                            int_count.append(int_index) # return the number of pandas column
                        int_count.sort()
                        int_counter = " ".join(map(str, int_count))
                        
                        if k not in column_count:
                            column_count.append(k) # return the number of pandas column
                        column_count.sort()
                        coll_count = " ".join(map(str, column_count))
                    print('Empty: '+str(counter)+' | Valid: '+str(anticounter)+' | Column = '+coll_count+' | Integer @ '+int_counter, end="\r", flush=True)
                    
                except pd.io.common.EmptyDataError:
                    counter = counter+1
                    # print('EmptyDataError: '+filename[46:], end="\n")
                    print('Empty: '+str(counter)+' | Valid: '+str(anticounter)+' | Column = '+coll_count+' | Integer @ '+int_counter, end="\r", flush=True)
                    file1.write(filename[74:]+"\n")
                    
                '''except ValueError:
                    pass'''
                    
    print('Empty: '+str(counter)+' | Valid: '+str(anticounter)+' | Column = '+coll_count+' | Integer @ '+int_counter, end="\n", flush=True)
    print('------------ Completed ------------')
    return

count_empty_dataframe("Heater Data")
count_empty_dataframe("Pressure Data")
count_empty_dataframe("RF Data")
```

# C. Filter common / missing files from the set

```python
folder_names = ["Heater Data", "Pressure Data", "RF Data", "Reports"]

rawnames = []
union_set = set()
folder_set = {}
diff = []
listw = {}
list_diff = {}
missing_filenames = {'Filenames':[]}
common_filenames = {'Filenames':[]}
filenames = list()
file_num = len(folder_names)
```

```python
''' Section A: Extract names of all files '''

for name in folder_names: missing_filenames.setdefault(name, [])

for folder_name in folder_names:  # iterate for each folder name
    base_path = r'C:\Users\Tanat\PycharmProjects\nhi-datacollection-ald\MachineSide\_Logfile'
    path = r''.join([base_path, '\\', folder_name])
    filenames.append(glob.glob(path + "/*"))

for k in range(0, file_num):  # iterate each folder
    for filename in filenames[k]:  # iterate each file in folder
        a = re.split('\\\\',filename)
        rawnames.append(a[-1][:-4])  # Extract only the file name, extension removed
    folder_set[k] = set(rawnames)  # Transfer list of file names to a separate set
    rawnames.clear()
```

```python
''' Section B: Create a list of missing files '''

for i in range(0, file_num):  # iterate each folder
    union_set = union_set | set(folder_set[i])  # Create a union set

for i in range(0, file_num):  # iterate each folder
    base = set(folder_set[i])  # copy each folder to a set variable, base
    list_diff = union_set.difference(base)  # find the relative complement of base set in union set 
                                            # This denotes all the files that base set is missing
    # for elem in list(list_diff):
    #     missing_filenames.setdefault(elem, []).append(folder_names[i])  
         # Appended here is the folder that does not have the filename
    
    for elem in list(list_diff):
        if elem not in missing_filenames.get('Filenames', ''):
            missing_filenames['Filenames'].append(elem)
            for name in folder_names: 
                missing_filenames[name].append("")
            missing_filenames[folder_names[i]][-1] = "Missing"
        else: 
            ndx = missing_filenames['Filenames'].index(elem)
            missing_filenames[folder_names[i]][ndx] = "Missing"
```

```python
''' Section C: Create a list of common files '''

intersection_list = set(folder_set[0])
for i in range(0, file_num):  # iterate each folder
    intersection_list = intersection_list.intersection(set(folder_set[i]))  # Create an intersection set
```

```python
''' Section D: Convert to dataframe '''    

df_missing = pd.DataFrame(missing_filenames) 
df_common = pd.DataFrame(intersection_list)

# Saving Pandas DataFrames
df_missing.to_csv("Results\missing files.csv", index=True, encoding='utf8')
df_common.to_csv("Results\common files.csv", index=True, encoding='utf8')

#(df_missing.style
#    .applymap(color_cells, subset=["Heater Data", "Pressure Data", "RF Data", "Reports"])
#    .set_table_styles(styles))

#(df_common.style
#    .set_table_styles(styles))
```

# D. Merge common & valid files 

```python
# Display and remove rows in heater data pandas
def clean_heater_file(filename):
    base_path =r'C:\Users\Tanat\PycharmProjects\nhi-datacollection-ald\venv\Scripts\Results\Heater Data (Fixed)'
    path = r''.join([base_path, '\\', filename, '.txt'])
    with open(path, "r") as file:
        data1 = pd.read_csv(file, sep="\t", engine='python')
        data1 = data1.iloc[:,1:-1]  # Remove the first column
        data1.rename(columns={'Heater Time':'Time'}, inplace=True)
        data1.rename(columns={'Cycles Remaining':'cycle1'}, inplace=True)
        data1.sort_values(by='Time')
    return data1
```

```python
# Display pressure data in pandas
def clean_pressure_file(filename):
    base_path =r'C:\Users\Tanat\PycharmProjects\nhi-datacollection-ald\MachineSide\_Logfile\Pressure Data'
    path = r''.join([base_path, '\\', filename, '.txt'])
    with open(path, "r") as file:
        data2 = pd.read_csv(file, sep="\t", engine='python')
        data2 = data2.iloc[:,1:]  # Remove the first column
        data2.rename(columns={'Pressure ':'Pressure'}, inplace=True)
        data2.rename(columns={'Pressure Time':'Time'}, inplace=True)
        data2.rename(columns={'Cycles Remaining':'cycle2'}, inplace=True)
        data2.sort_values(by='Time')
    return data2
```

```python
# Display RF data in pandas
def clean_rf_file(filename):
    base_path =r'C:\Users\Tanat\PycharmProjects\nhi-datacollection-ald\MachineSide\_Logfile\RF Data'
    path = r''.join([base_path, '\\', filename, '.txt'])
    with open(path, "r") as file:
        data3 = pd.read_csv(file, sep="\t", engine='python')
        data3 = data3.iloc[:,1:]  # Remove the first column
    return data3
    #data3.hist(column='Cycles Remaining')
```

```python
def merge_heater_pressure_files(data1, data2):
    data1 = data1.replace({pd.np.nan: ''})
    data2 = data2.replace({pd.np.nan: ''})
    data = pd.merge_asof(data1, data2, on='Time')
    data = data[1:].replace({pd.np.nan: ''})  # removes the first line & replace Nan
    return data
    
    #list(data.columns.values)
```

```python
def show_cycle_histogram(filename, data):
    #data.hist(column='cycle1')  # Heater data
    #data.hist(column='cycle2')  # Pressure data
    
    # Below is to print the histogram bins and frequencies
    A = list(data['cycle1'])
    count1, division1 = np.histogram(A[1:])
    B = list(data['cycle2'])
    count2, division2 = np.histogram(B[1:])
    if str(division1) == str(division2):
        data1 = pd.DataFrame({'division': division1[:-1], 'count_heater': count1, 'count_pressure': count2})
    else: 
        print("division not the same")
    #    print(str(division1))
    #    print(str(count1))
    #    print(str(division2))
    #    print(str(count2))

    #(data1.style
    #          .set_table_styles(styles))

    plt.plot( division1[:-1] , count1, marker='', markerfacecolor='blue', markersize=5, color='skyblue', linewidth=2, label="Heater")
    plt.plot( division2[:-1] , count2, marker='', color='olive', linewidth=2, linestyle='dashed', label="Pressure")
    plt.title('Cycle Remaining Histogram\n'+filename)
    plt.savefig('Results/figure/CycleHistogram_'+filename+'.png')
    plt.close('all')

    return
```

```python
## Find a diff way to save figure
# so that you can post-process it 

def timeseries_plot(filename, data):
    # filter by timespan of ALD run or file size: os.path.getsize(filename) > 5000 (bytes)
    #if average(float(list(data['Time'])[0:5])) > 60.0 or cloat(data['Time'][-1]) < 300.0:
    #    return
    
    
    ''' Putting lower limit to the time length of record we will look at '''
    data['Time'] = data['Time'].apply(lambda x: x/60)
    if data['Time'][data.index[-1]] < 5.0:
        # raise ValueError("Too short of a record")    
        return 
    
    
    ''' Calculate moving averages for heater temperature and pressure '''
    plot_header = list()
    num_heater = sum(1 for word in data.columns.tolist() if 'Heater' in word)
    for i in range(1,num_heater+1):
        header = 'Heater '+str(i)
        new_header = header+' MA'
        data[new_header] = data[header].rolling(window=1).mean()
        plot_header.append(new_header)
    data['Pressure MA'] = data['Pressure'].rolling(window=1).mean()

    
    ''' Plotting the combined Pressure / Heater Data '''
    # plt.clf()  # Clear an entire figure, use when trying to plot new info on the same plot
    ax1 = data.plot(x='Time', y=plot_header \
                    , figsize=(20,14), fontsize=16)
    ax1.legend(prop={'size':20}, bbox_to_anchor=(1.2,0.9))
    ax1.set_xlim(0,1)  # Limit to only the first 50 minutes
    # ax1.yaxis.set_major_locator(ticker.MultipleLocator(1))
    plt.ylim(50, 250)        
    plt.xlabel('Time (mins)', size = 20)
    plt.ylabel('Temperature (°C)', size = 20)
    

    ax2 = ax1.twinx()
    ax2.spines['right'].set_position(('axes', 1.0))
    data.plot(ax=ax2, x='Time', y='Pressure MA', fontsize=16)
    ax2.legend(prop={'size':20}, bbox_to_anchor=(1.2,1))
    plt.ylabel('Pressure (Torr)', size = 20)
    plt.title('Time Series: Pressure & Heater Data', size = 26)
    ax2.set_xlim(0,1)  # Limit to only the first 50 minutes
    
    
    ''' Save the figure into PNG file '''
    plt.savefig('Results/figure/'+filename+'.png')
    # plt.close('all')
    return
```

```python
def add_value_labels(ax, spacing=5):
    """Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """

    # For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = spacing
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:.1f}".format(y_value)

        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(0, space),          # Vertically shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            ha='center',                # Horizontally center label
            va=va)                      # Vertically align label differently for
                                        # positive and negative values.
```
